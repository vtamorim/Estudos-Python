```python
import matplotlib.pyplot as plt
import numpy as np


tempos = [0.0, 0.0613, 0.2580, 0.3952, 0.4523]
velocidades = [0.0, 1.436, 3.054, 4.372, 4.904]
posicoes = [0.0, 0.044, 0.394, 0.864, 1.109]

t_continuo = np.linspace(0, 0.5, 100)
g_estimado = 9.8


plt.figure(figsize=(8, 5))
plt.plot(t_continuo, g_estimado * t_continuo, '--', color='gray', label='Ajuste Teórico (v = g.t)')
plt.scatter(tempos, velocidades, color='red', zorder=5, label='Dados Experimentais')
for i, txt in enumerate(velocidades):
    plt.annotate(f'({tempos[i]:.4f}, {txt})', (tempos[i], velocidades[i]), textcoords="offset points", xytext=(10,-5), ha='left')
plt.title('Velocidade em função do tempo')
plt.xlabel('Tempo t (s)')
plt.ylabel('Velocidade v (m/s)')
plt.grid(True, linestyle=':', alpha=0.7)
plt.legend()
plt.show()

plt.figure(figsize=(8, 5))
plt.plot(t_continuo, 0.5 * g_estimado * t_continuo**2, '--', color='gray', label='Ajuste Teórico (s = 0.5.g.t²)')
plt.scatter(tempos, posicoes, color='blue', zorder=5, label='Dados Experimentais')
for i, txt in enumerate(posicoes):
    plt.annotate(f'({tempos[i]:.4f}, {txt})', (tempos[i], posicoes[i]), textcoords="offset points", xytext=(10,-5), ha='left')
plt.title('Posição em função do tempo')
plt.xlabel('Tempo t (s)')
plt.ylabel('Posição d (m)')
plt.grid(True, linestyle=':', alpha=0.7)
plt.legend()
plt.show()
```
